```python
# Cell [18]
import numpy as np

# Apply Log Transformation (Log1p handles 0 values safely)
df['rfh_log'] = np.log1p(df['rfh'])

print("Log transformation applied to squash extreme peaks while keeping the data.")

# Cell [19]
# 1. Extract Month and Dekad
df['month'] = df['date'].dt.month
df['dekad'] = df['date'].dt.day.map({1:1, 11:2, 21:3})

# 2. Convert to Cyclical coordinates
df['month_sin'] = np.sin(2 * np.pi * df['month']/12)
df['month_cos'] = np.cos(2 * np.pi * df['month']/12)

df['dekad_sin'] = np.sin(2 * np.pi * df['dekad']/3)
df['dekad_cos'] = np.cos(2 * np.pi * df['dekad']/3)

# Cell [20]
# Sort again just to be safe
df = df.sort_values(['adm_id', 'date']).reset_index(drop=True)

# 1. Create Lags (Rain from previous periods)
df['rfh_lag1'] = df.groupby('adm_id')['rfh_log'].shift(1)
df['rfh_lag3'] = df.groupby('adm_id')['rfh_log'].shift(3)

# 2. Create Rolling Mean (Trend of Last 3 dekads)
df['rfh_roll3_mean'] = df.groupby('adm_id')['rfh_log'].transform(lambda x: x.rolling(window=3).mean().shift(1))

# Drop the rows where we don't have enough history yet (the NaN rows created by shifts)
df = df.dropna().reset_index(drop=True)

# Cell [21]
from sklearn.preprocessing import MinMaxScaler
import joblib

# 1. Define Features to use
features = ['rfh_avg', 'rfq', 'month_sin', 'month_cos', 'dekad_sin', 'dekad_cos',
            'rfh_lag1', 'rfh_lag3', 'rfh_roll3_mean']

X = df[features].values
y = df['rfh_log'].values.reshape(-1, 1)

# 2. Scale the data
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# 3. Save the scalers (Very important for the Web App Later!)
joblib.dump(scaler_X, 'scaler_X.pkl')
joblib.dump(scaler_y, 'scaler_y.pkl')
```

```python
from sklearn.model_selection import train_test_split

# 1. Split the data into Training and Testing sets (e.g., 80% train, 20% test)
X_train_raw, X_test_raw, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# 2. Reshape input to be 3D [samples, time steps, features] for LSTM
# Your model summary shows a sequence length (time steps) of 1
X_train = X_train_raw.reshape((X_train_raw.shape[0], 1, X_train_raw.shape[1]))
X_test = X_test_raw.reshape((X_test_raw.shape[0], 1, X_test_raw.shape[1]))

print(f"X_train shape: {X_train.shape}") # Should be (samples, 1, 9)
print(f"y_train shape: {y_train.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization, Input

# 1. Build the Brain (Architecture)
model = Sequential([
    # We use '12' features now because of your advanced feature engineering
    Input(shape=(X_train.shape[1], X_train.shape[2])),
    
    # Layer 1: High capacity LSTM to Learn the big rain cycles
    LSTM(128, activation='tanh', return_sequences=True),
    BatchNormalization(),
    Dropout(0.2),
    
    # Layer 2: Refines the patterns
    LSTM(64, activation='tanh'),
    Dropout(0.2),
    
    # Dense Layers to calculate the final mm number
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1) # Final prediction
])

# 2. Compile the model
# Huber Loss is great here because it helps with those extreme peaks you found
model.compile(optimizer='adam', loss='huber', metrics=['mae'])

model.summary()
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping

# 1. Slows down learning if accuracy stops improving (helps find the peaks)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001)

# 2. Stops training if the model starts to overfit
early_stop = EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True)

print("✅ Callbacks and Model are ready. Now you can run the training cell!")
```

```python
import numpy as np

# 1. Start with a weight of 1 for all training samples
weights = np.ones(len(y_train))

# 2. Identify the "peaks" (high rainfall values)
# Since y_train is scaled between 0 and 1, we look for values in the upper range
# We'll give these peaks a weight of 10 so the model focuses on them
weights[y_train.flatten() > 0.5] = 10 

print(f"Weights calculated: {len(weights)} samples ready.")
print(f"Number of peak samples receiving 10x importance: {np.sum(weights == 10)}")
```

```python
# Start the Learning process
history = model.fit(
    X_train, y_train,
    epochs=100,             # We use 100 but EarlyStopping will likely stop it sooner
    batch_size=32,
    validation_split=0.1,
    sample_weight=weights,  # THIS IS THE KEY: Forces AI to learn the 140mm peaks
    callbacks=[reduce_lr, early_stop],
    verbose=1
)

print("🏆 Training Complete! Your model is now a professional weather expert.")
```
